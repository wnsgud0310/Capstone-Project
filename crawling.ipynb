```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# 웹 드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 네이버 중고나라 카페 접속
driver.get("https://cafe.naver.com/joonggonara")

# 검색어 입력 및 검색 실행
search_box = driver.find_element(By.ID, "topLayerQueryInput")
search_box.send_keys("아이폰")
search_box.send_keys(Keys.RETURN)

# 동적 페이지 로딩 대기
time.sleep(5)

# 프레임 변경
driver.switch_to.frame("cafe_main")

# 목록 크기 선택 (드롭다운 메뉴 활성화)
list_size_button = driver.find_element(By.ID, "listSizeSelectDiv")
list_size_button.click()
time.sleep(2)  # 드롭다운 메뉴 로딩 대기

# "50개씩" 선택
fifty_per_page_option = driver.find_element(By.CSS_SELECTOR, "#listSizeSelectDiv > ul > li:nth-child(7) > a")
fifty_per_page_option.click()
time.sleep(3)  # 새 목록 크기로 페이지 로딩 대기

# 데이터 수집
all_posts = []
current_page = 1
max_pages = 100  # 최대 페이지 수 설정

while current_page <= max_pages:
    posts = driver.find_elements(By.CSS_SELECTOR, ".article-board > table > tbody > tr")
    for post in posts:
        try:
            title = post.find_element(By.CSS_SELECTOR, ".article").text
            link = post.find_element(By.CSS_SELECTOR, "a.article").get_attribute('href')
            all_posts.append((title, link))
            print(title, link)
        except Exception as e:
            print("Error processing post:", e)

    # 다음 페이지 번호 클릭
    if current_page < max_pages:
        page_link_index = 3 + (current_page % 10)
        try:
            # '다음' 버튼 클릭 시
            if current_page % 10 == 0:
                next_page_button = driver.find_element(By.CSS_SELECTOR, ".prev-next .pgR")
                next_page_button.click()
            else:
                # 현재 페이지 블록 내에서 다음 페이지 클릭
                next_page_link = driver.find_element(By.CSS_SELECTOR, f".prev-next a:nth-child({page_link_index})")
                next_page_link.click()
            time.sleep(5)  # 페이지 로딩 대기
        except Exception as e:
            print("No more pages or error:", e)
            break

    current_page += 1

# 웹 드라이버 종료

driver.quit()

import pandas as pd

# 수집한 데이터를 데이터프레임으로 변환
df = pd.DataFrame(all_posts, columns=['Title', 'Link'])


# CSV 파일로 저장
df.to_csv('iphone.csv', index=False, encoding='utf-8-sig')
```
